1. Outliers are points in a data set which varies significantly from the other data points .

2. The no. of data points which is falling between 25th percentile and 75th percentile is known as "Inter-quartile range" .  

```python
freq, bins = pd.cut(df["Marks"], bins=5, include_lowest=True, right=False, retbins=True)
freq_table = freq.value_counts().sort_index()
mid_points = [(bins[i] + bins [i+1]) / 2 for i in range(len(bins)-1)]
plt.plot(mid_points, freq_table, marker='o')
plt.xlabel("Class Midpoints")
plt.ylabel("Frequency")
plt.title("Frequency Polygon")
plt.show()
```

```python
cum_freq = freq_table.cumsum()

plt.plot(mid_points, cum_freq, marker='o')
plt.xlabel("Class Midpoints")
plt.ylabel("Cumulative Frequency")
plt.title("Ogive (Cumulative Frequency Curve)")
plt.show()
```

The average temp. of the residential area is 28 degree C per unit of electricity is costing $1 . Calculate the total electricity bill that tarun has to pay .the ratingt of each bulb is 60 W , ac is 1.5 kw and cctv is 0.5 kw . Each equipment is used for 8 hours per day and 30 days . Calculate for the month.

```python
# Given data
hours_per_day = 8
days_per_month = 30
total_hours = hours_per_day * days_per_month
cost_per_unit = 1  # $ per kWh

# Light bulbs
num_bulbs = 15 * 10
bulb_power_kw = 60 / 1000  # convert W to kW
bulb_energy = num_bulbs * bulb_power_kw * total_hours
bulb_cost = bulb_energy * cost_per_unit

# Air conditioners
num_acs = 10 * 2
ac_power_kw = 1.5
ac_energy = num_acs * ac_power_kw * total_hours
ac_cost = ac_energy * cost_per_unit

# CCTV equipment
num_cctv = 250
cctv_power_kw = 0.5
cctv_energy = num_cctv * cctv_power_kw * total_hours
cctv_cost = cctv_energy * cost_per_unit

# Total calculations
total_energy = bulb_energy + ac_energy + cctv_energy
total_cost = bulb_cost + ac_cost + cctv_cost

# Output
print("Monthly Electricity Consumption (kWh):", total_energy)
print("Total Monthly Electricity Bill ($):", total_cost)
```

A telecom company wants to analysis coustmer behaviour.To improve its sevice quality and revenue
1. barchart plan type
2. histogram for bill amt.
3. boxplot data usage
4. line graph data usage
5. mean , median and mode for bill amt

7. range, standard deviation, IQR ,pierson corelation  ,spearman corelation
8. probability of high bill(bill amt > 800)
9. emperical probabilty for no. of complains
10. Predict the bill amt for Tarun who has used 120 gb of data usage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# -------------------------------
# 1. Create Sample Dataset
# -------------------------------
np.random.seed(42)

df = pd.DataFrame({
    "data_usage": np.random.normal(10, 3, 40).round(2),      # GB/day
    "bill_amount": np.random.normal(700, 150, 40).round(2), # INR
    "complaints": np.random.poisson(2, 40),
    "gender": np.random.choice(["Male", "Female"], 40),
    "plan_type": np.random.choice(["Basic", "Standard", "Premium"], 40),
    "satisfaction": np.random.randint(1, 6, 40)
})

print("Sample Data:")
print(df.head())

# -------------------------------
# 2. Bar Chart  Plan Type
# -------------------------------
df['plan_type'].value_counts().plot(kind='bar')
plt.title("Distribution of Plan Types")
plt.xlabel("Plan Type")
plt.ylabel("Number of Customers")
plt.show()

# -------------------------------
# 3. Histogram  Bill Amount
# -------------------------------
plt.hist(df['bill_amount'], bins=10)
plt.title("Histogram of Bill Amount")
plt.xlabel("Bill Amount (INR)")
plt.ylabel("Frequency")
plt.show()

# -------------------------------
# 4. Box Plot  Data Usage
# -------------------------------
plt.boxplot(df['data_usage'])
plt.title("Box Plot of Data Usage")
plt.ylabel("Data Usage (GB/day)")
plt.show()

# -------------------------------
# 5. Line Graph  Data Usage
# -------------------------------
plt.plot(df['data_usage'])
plt.title("Line Graph of Data Usage")
plt.xlabel("Customer Index")
plt.ylabel("Data Usage (GB/day)")
plt.show()

# -------------------------------
# 6. Mean, Median, Mode  Bill Amount
# -------------------------------
mean_bill = df['bill_amount'].mean()
median_bill = df['bill_amount'].median()
mode_bill = df['bill_amount'].mode()[0]

print("\nBill Amount Statistics:")
print("Mean:", round(mean_bill, 2))
print("Median:", round(median_bill, 2))
print("Mode:", round(mode_bill, 2))

# -------------------------------
# 7. Weighted Mean  Bill Amount w.r.t Data Usage
# -------------------------------
weighted_mean = np.average(df['bill_amount'], weights=df['data_usage'])
print("\nWeighted Mean Bill Amount:", round(weighted_mean, 2))

# -------------------------------
# 8. Range, Std Dev, IQR
# -------------------------------
range_bill = df['bill_amount'].max() - df['bill_amount'].min()
std_bill = df['bill_amount'].std()
iqr_bill = stats.iqr(df['bill_amount'])

print("\nDispersion Measures:")
print("Range:", round(range_bill, 2))
print("Standard Deviation:", round(std_bill, 2))
print("IQR:", round(iqr_bill, 2))

# -------------------------------
# 9. Pearson & Spearman Correlation
# -------------------------------
pearson_corr = df['data_usage'].corr(df['bill_amount'], method='pearson')
spearman_corr = df['data_usage'].corr(df['bill_amount'], method='spearman')

print("\nCorrelation:")
print("Pearson Correlation:", round(pearson_corr, 3))
print("Spearman Correlation:", round(spearman_corr, 3))

# -------------------------------
# 10. Probability of High Bill (> 800)
# -------------------------------
high_bill_prob = (df['bill_amount'] > 800).mean()
print("\nProbability of High Bill (>800 INR):", round(high_bill_prob, 3))

# -------------------------------
# 11. Empirical Probability  Number of Complaints
# -------------------------------
complaint_prob = df['complaints'].value_counts(normalize=True)
print("\nEmpirical Probability of Complaints:")
print(complaint_prob)

# -------------------------------
# 12. Predict Bill Amount for Tarun (120 GB Usage)
# Using Linear Regression
# -------------------------------
slope, intercept, r, p, std_err = stats.linregress(df['data_usage'], df['bill_amount'])

tarun_data_usage = 120
predicted_bill = intercept + slope +5 * tarun_data_usage

print("\nPrediction:")
print(f"Predicted Bill Amount for Tarun (120 GB usage): INR {round(predicted_bill, 2)}")
```

Central limit theorem(CLT) :- If we take many samples from any population and calulate there statistical mean then the mean values will form a bell shaped curve (normal distribution) as long as the sample size is large.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed (100)
population_size = 10000
population = pd.DataFrame({
    "id": range (population_size),
    "income": np.random.normal(50000, 15000,
    population_size),
    "gender": np.random.choice (["Male", "Female"], population_size, p=[0.6, 0.4]),
    "city": np.random.choice(["Kolkata", "Delhi", "Mumbai"], population_size)
})

population.head()
```

```python
sample_means = []
for _ in range(1000):
    sample = population["income"].sample (n = 50)
    sample_means.append(sample.mean())
plt.hist(sample_means, bins=30)
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean Income")
plt.ylabel("Frequency")
plt.show()
```

```python
random_sample = population.sample ( n = 500 , random_state=42)
print("Random Sample Mean Income:", random_sample["income"].mean())
```

```python
stratified_sample = population.groupby("gender", group_keys=False).apply(
lambda x: x.sample(frac=500/population_size, random_state=42)
)
print("Stratified Sample Distribution:")
print(stratified_sample["gender"].value_counts())
```

```python
k = population_size // 500
systematic_sample = population.iloc[::k]
print("Systematic Sample Size:", len(systematic_sample))
```

```python
selected_cities = np.random.choice (population ["city"].unique(), size=1)
cluster_sample = population [population ["city"].isin (selected_cities)]
print("Cluster Sample City:", selected_cities)
print("Cluster Sample Size:", len (cluster_sample))
```

```python
sample = population["income"].sample(n=100)
sample_std = sample.std()
n = len(sample)
standard_error = sample_std / np.sqrt(n)
print("Sample Standard Deviation:", sample_std)
print("Standard Error:", standard_error)
```

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate a highly skewed population
population = np.random.exponential(scale=10, size=100000)

# Sample sizes
sample_sizes = [10, 30, 100, 300]

print("Sample Size\tStandard Deviation")
for n in sample_sizes:
    sample = np.random.choice(population, size=n, replace=False)
    sd = np.std(sample, ddof=1)  # sample standard deviation
    print(f"{n}\t\t{sd:.2f}")
```

```python
# Advertising Spend vs Sales - Statistical Analysis
# Author: Surya Pratap Singh

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# -----------------------------
# 1. Create Dataset
# -----------------------------
advertising_spend = np.array([10, 15, 20, 25, 30, 35, 40])
sales = np.array([40, 55, 70, 85, 100, 115, 130])

data = pd.DataFrame({
    'Advertising_Spend': advertising_spend,
    'Sales': sales
})

print("Dataset:\n", data)

# -----------------------------
# 2. Descriptive Statistics
# -----------------------------
print("\nDescriptive Statistics:")
print(data.describe())

# -----------------------------
# 3. Scatter Plot
# -----------------------------
plt.figure()
plt.scatter(advertising_spend, sales)
plt.xlabel("Advertising Spend ( in 1000s)")
plt.ylabel("Sales (Units in 1000s)")
plt.title("Advertising Spend vs Sales")
plt.show()

# -----------------------------
# 4. Correlation Analysis
# -----------------------------
correlation = np.corrcoef(advertising_spend, sales)[0, 1]
print("\nCorrelation Coefficient:", correlation)

# -----------------------------
# 5. Linear Regression
# -----------------------------
slope, intercept, r_value, p_value, std_err = stats.linregress(advertising_spend, sales)

print("\nLinear Regression Equation:")
print(f"Sales = {slope:.2f} * Advertising_Spend + {intercept:.2f}")

# Regression line
predicted_sales = slope * advertising_spend + intercept

plt.figure()
plt.scatter(advertising_spend, sales, label="Actual Data")
plt.plot(advertising_spend, predicted_sales, label="Regression Line")
plt.xlabel("Advertising Spend ( in 1000s)")
plt.ylabel("Sales (Units in 1000s)")
plt.title("Linear Regression Model")
plt.legend()
plt.show()

# -----------------------------
# 6. Hypothesis Testing
# -----------------------------
# H0: Advertising spend has no effect on sales
# H1: Advertising spend significantly affects sales

alpha = 0.05
print("\nHypothesis Testing:")
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis: Advertising spend significantly affects sales.")
else:
    print("Fail to reject the null hypothesis: No significant effect found.")

# -----------------------------
# 7. Prediction
# -----------------------------
new_ad_spend = 45
predicted_value = slope * new_ad_spend + intercept
print(f"\nPredicted sales for {new_ad_spend}k advertising spend:", round(predicted_value, 2))
```
